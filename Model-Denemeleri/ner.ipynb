```python
from transformers import AutoModelForTokenClassification, AutoTokenizer, pipeline
```

```python
task = "ner"
model_checkpoint = "xlm-roberta-base"
batch_size = 8 
label_list = ['O', 'B-PER', 'I-PER', 'B-ORG', 'I-ORG', 'B-LOC', 'I-LOC']
max_length = 512 
learning_rate = 2e-5 
num_train_epochs = 2 
weight_decay = 0.01
```

```python
model = AutoModelForTokenClassification.from_pretrained("akdeniz27/xlm-roberta-base-turkish-ner")
```

```python
tokenizer = AutoTokenizer.from_pretrained("akdeniz27/xlm-roberta-base-turkish-ner")
```

```python
ner = pipeline('ner', model=model, tokenizer=tokenizer, aggregation_strategy="simple")
a = ner("9 yaşındaki bir çocuğun yaptığı bir seçimle yaşayabileceği farklı hayatları konu alan Leonardo Dicaprio'un Londra'da oynadığı film.")
```
