# Search Fonksiyonu

```python
def search(querry, description_embeddings, model, movie_names, k):
    querry_embedding = model.encode(querry)
    results = np.array(util.cos_sim(querry_embedding, description_embeddings)).reshape(-1)
    a = zip(movie_names, results)
    result_list = sorted(a, key = lambda x: x[1])[::-1]
    for i in result_list[:k]:
        print(i[0], ": ", i[1])
```

# Denemeler ve sonuçlar

```python
test_df = pd.read_csv("form.csv")
test_df
```

Model skoru ölçme fonksiyonu

```python
def model_score(model, corpus_embeddings, movie_names, test_df, k):
    count = 0
    for idx in range(len(test_df["Film İsmi"])):
        name = test_df["Film İsmi"][idx]
        desc = test_df["Film Hakkında"][idx]
        querry_embedding = model.encode(desc)
        results = np.array(util.cos_sim(querry_embedding, corpus_embeddings)).reshape(-1)
        a = zip(movie_names, results)
        result_list = sorted(a, key = lambda x: x[1])[::-1]
        result_list_names = [i[0] for i in result_list[:k]]
        if name in result_list_names:
            count += 1
    
    return count/len(test_df.index)
```
